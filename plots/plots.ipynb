۱. نمودار توزیع (هیستوگرام)

  * کاربرد: برای نمایش توزیع یک متغیر پیوسته استفاده می‌شود.

  * دلیل: به درک پراکندگی و میانگین مرکزی داده‌ها کمک می‌کند.

  * شخصی‌سازی: می‌توانید رنگ‌ها را سفارشی کنید، تخمین چگالی هسته‌ای (kde) اضافه کنید، اندازه باکس‌ها را تنظیم کنید و نمودار‌های rug را اضافه کنید.

۲. نمودار گروه‌بندی (نمودار میله‌ای)

  * کاربرد: مناسب برای نمایش تعداد یا مقادیر جمع‌شده متغیرهای دسته‌ای است.

  * دلیل: برای مقایسه دسته‌ها و نمایش توزیع‌های فراوانی مفید است.

  * شخصی‌سازی: گزینه‌های شخصی‌سازی شامل تغییر رنگ‌های نمودار، اضافه کردن اشکال خطای (ci)، تنظیم عرض نمودار میله‌ای و تغییر ترتیب دسته‌ها است.

۳. نمودار پراکندگی

  * کاربرد: نشان دادن رابطه بین دو متغیر پیوسته است.

  * دلیل: برای تشخیص الگوها، روندها و داده‌های نادرست مفید است.

  * شخصی‌سازی: می‌توانید اشکال نشانگر، رنگ‌ها، اندازه‌ها، اضافه کردن خطوط روند (regplot) و تنظیم محدوده محورها را تغییر دهید.

۴. نمودار جفتی

  * کاربرد: نمایش روابط دو به دو بین چندین متغیر در مجموعه داده است.

  * دلیل: برای کاوش در روابط و شناسایی الگوها در داده‌های بالابعدادی مفید است.

  * شخصی‌سازی: می‌توانید نشانگرهای نقطه‌ای، هیستوگرام‌ها/تخمین‌های چگالی هسته‌ای قطره‌ای و نقشه رنگ/گرادیان را تنظیم کنید.
    
۵. نمودار حرارتی

  * کاربرد: نمایش ماتریس همبستگی یا هر مجموعه داده ۲ بعدی دیگر.

  * دلیل: برای شناسایی الگوها، همبستگی‌ها یا خوشه‌بندی در داده‌های جدولی مفید است.

  * شخصی‌سازی: گزینه‌های شخصی‌سازی شامل تغییر رنگ‌های نقشه (cmap)، اضافه کردن حاشیه‌نویسی‌ها (annot)، تنظیم اندازه سلول و تغییر برچسب‌های محور است.

۶.  : نمودار رگرسیون

  * کاربرد: نمایش رابطه بین دو متغیر پیوسته به همراه خط رگرسیون.

  * دلیل: برای مشاهده روابط خطی و پیش‌بینی نتایج مفید است.

  * شخصی‌سازی: می‌توانید اشکال نشانگر، رنگ‌ها، اندازه‌ها، اضافه کردن بازه‌های اطمینان (ci) و تنظیم محدوده محورها را تغییر دهید.

۷. نمودار مشترک

  * کاربرد: ترکیب نمودارهای یکتا و دوتایی برای نمایش روابط بین دو متغیر.

  * دلیل: برای کاوش در توزیع‌های مشترک، توزیع‌های لبه‌ای و همبستگی‌ها مفید است.

  * شخصی‌سازی: گزینه‌های شخصی‌سازی شامل تغییر انواع نمودار (kind)، اشکال نشانگر، رنگ‌ها، اندازه‌ها و اضافه کردن حاشیه‌نویسی‌ها است.


# استفاده از کتابخانه panda

```python
# وارد کردن کتابخانه
import pandas as pd

# ساخت یک داده‌فریم ساده
data = {'Name': ['Alise', 'Bob', 'Clark', 'David', 'Elen', 'Ferank'],
        'Age': [25, 30, 35, 40, 45, 50],
        'Score': [70, 85, 82, 90, 88, 95]}

df = pd.DataFrame(data)
```

```python
df.plot(kind='line', x='Name', y='Age', marker='o', color='b', linestyle='-', figsize=(10, 6))
```

```python
df.plot(kind='bar', x='Name', y='Score', color='g', figsize=(10, 6))
```

```python
df.plot(kind='barh', x='Name', y='Score', color='orange', figsize=(10, 6))
```

```python
df.plot(kind='pie', y='Age', labels=df['Name'], autopct='%1.1f%%', figsize=(10, 6))
```

```python
df.plot(kind='scatter', x='Age', y='Score', color='purple', s=100, figsize=(10, 6))
```

```python
df.plot(kind='box', y='Age', figsize=(10, 6))
```

# استفاده از کتابخانه  ی matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample Data
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
# Line Plot
plt.plot(x, y, color='blue')
plt.title('Line Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# Scatter Plot
plt.scatter(x, y, color='red')
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# Bar Plot
plt.bar(x[:10], y[:10], color='green')
plt.title('Bar Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# Histogram
plt.hist(y, bins=20, color='orange')
plt.title('Histogram')
plt.xlabel('Y')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box Plot
plt.boxplot(y)
plt.title('Box Plot')
plt.show()
```

```python
# Pie Chart
sizes = [20, 30, 40, 10]
plt.pie(sizes, labels=['A', 'B', 'C', 'D'], autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
# Heatmap
data = np.random.rand(10, 10)
plt.imshow(data, cmap='coolwarm', interpolation='nearest')
plt.title('Heatmap')
plt.show()
```

# استفاده از کتابخانه seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample Data
tips = sns.load_dataset("tips")
```

```python

# Distribution Plot (Histogram)
sns.histplot(tips['total_bill'], kde=True, color='skyblue')
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

```python
# Categorical Plot (Bar Plot)
sns.countplot(x='day', data=tips, palette='pastel')
plt.title('Count of Observations by Day')
plt.xlabel('Day')
plt.ylabel('Count')
plt.show()
```

```python
# Scatter Plot
sns.scatterplot(x='total_bill', y='tip', data=tips, color='green')
plt.title('Tip vs Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
# Pair Plot
sns.pairplot(tips, hue='sex', palette='husl')
plt.title('Pair Plot')
plt.show()
```

```python
# Exclude Categorical Variables
numeric_tips = tips.select_dtypes(include=['float64', 'int64'])

# Calculate Correlation Matrix
correlation_matrix = numeric_tips.corr()

# Plot Heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python

# Regression Plot
sns.regplot(x='total_bill', y='tip', data=tips, color='purple')
plt.title('Regression Plot: Tip vs Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python

# Joint Plot
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex', color='orange')
plt.title('Joint Plot: Tip vs Total Bill')
plt.show()
```
